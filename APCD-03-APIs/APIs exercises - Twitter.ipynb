Define the csv file and the headers to store. By default, you will save user, text and hashtags.

```python
csvres_headers =['user', 'text', 'hashtags']

csvfile = open('SevillaTweets.csv', 'w')
writer = csv.writer(csvfile, delimiter='|')

writer.writerow(csvres_headers)
```

For each tweet, replace the user for the word USER and the retweets for the word RETWEET. Also, get the hashtags. Finally, save the info in the csv file.

```python
for tweet in results['statuses']:
    row = []
    hashtags = []
    tweet['user']='USER'
    if 'retweeted_status' in tweet.keys():
        tweet['text']='RETWEET'
    for hashtag in tweet['entities']['hashtags']:
        hashtags.append(hashtag['text'])
    row = [tweet['user'],tweet['text'],hashtags]
    writer.writerow(row)

csvfile.close()    
```
